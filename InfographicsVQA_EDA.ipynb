```python
ANSWER = []
EVIDENCE = []
OPERATION = []
for i,l in enumerate(json_valid['data']):
    try:
        ANSWER.append(l['answer_type'][0])
        EVIDENCE.append(l['evidence'][0])
        OPERATION.append(l['operation/reasoning'][0])
    except:
        pass
from collections import Counter
answer_dist = Counter(ANSWER)
evi_dist = Counter(EVIDENCE)
oper_dist = Counter(OPERATION)
```

```python
answer_total = sum(answer_dist.values())
evi_total = sum(evi_dist.values())
oper_total = sum(oper_dist.values())
```

```python
for i,l in answer_dist.items():
    percentaget = (l/answer_total)*100
    print(f'{i}: {percentaget:.2f}%')
answer_dist
    
```

```python
for i,l in evi_dist.items():
    percentaget = (l/evi_total)*100
    print(f'{i}: {percentaget:.2f}%')
evi_dist
```

```python
for i,l in oper_dist.items():
    percentaget = (l/oper_total)*100
    print(f'{i}: {percentaget:.2f}%')
oper_dist
```
