```python
# SP 데이터 정렬 → 500장 추출

with open('SP_val_sorted.json', 'r', encoding='utf-8') as f:
    data3 = json.load(f)

stack_list = []
data_list = []
length = 0
idx = 0

data_select = data3

while length < 500:  # 이미지 개수 지정
    data_list.append(data_select['data'][idx])
    if data_select['data'][idx]['image'] not in stack_list:
        stack_list.append(data_select['data'][idx]['image'])
        length += 1
    idx += 1

form = {
        'dataset_name' : data_select['dataset_name'],
        'dataset_version' : data_select['dataset_version'],
        "dataset_split": data_select["dataset_split"],
        'data' : data_list
}

with open('SP_val_select.json', 'w', encoding='utf-8') as f:
        json.dump(form, f, indent=4)

################## 점검 #########################
stack = []
with open('SP_val_select.json', 'r', encoding='utf-8') as f:
    data_select = json.load(f)
for i in data_select['data']:
    if i['image'] not in stack:
        stack.append(i['image'])
print('image 개수 : ', len(stack))
print('question 개수 : ',len(data_select['data']))
```

```python
# SP image, json 옮기기

import os
import shutil

destination_img_folder = "../data/task1_SP/SP_image_select"

if not os.path.exists(destination_img_folder):
    os.makedirs(destination_img_folder)

with open('SP_val_select.json', 'r', encoding='utf-8') as f:
    data_select = json.load(f)

data_list = data_select.get("data", [])

for data_entry in data_list:
    image_name = data_entry.get("image", "")[10:]
    if image_name:
        source_img_path = os.path.join("../data/task1_SP/documents", image_name)
        destination_img_path = os.path.join(destination_img_folder, image_name)
        if os.path.exists(source_img_path):
            shutil.copy2(source_img_path, destination_img_path)

print('image 개수 : ', len(os.listdir(destination_img_folder)))

###############################################################

destination_ocr_folder = "../data/task1_SP/SP_ocr_select"

if not os.path.exists(destination_ocr_folder):
    os.makedirs(destination_ocr_folder)

with open('SP_val_select.json', 'r', encoding='utf-8') as f:
    data_select = json.load(f)

data_list = data_select.get("data", [])

for data_entry in data_list:
    ocr_name = data_entry.get("image", "")[10:][:-4] + '.json'
    if ocr_name:
        source_ocr_path = os.path.join("../data/task1_SP/ocr_results", ocr_name)
        destination_ocr_path = os.path.join(destination_ocr_folder, ocr_name)
        if os.path.exists(source_ocr_path):
            shutil.copy2(source_ocr_path, destination_ocr_path)

print('ocr 개수 : ', len(os.listdir(destination_ocr_folder)))
```
