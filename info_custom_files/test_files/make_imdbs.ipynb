```python
import json
import numpy as np
from PIL import Image
```

```python
with open('/home/ges/level3-cv-productserving-cv-10/data/infographics/qas/infographicsVQA_val_v1.0_withQT.json', 'r') as file:
    data = json.load(file)
data
```

```python
imdb_data = []

i = 0

for item in data['data']:
    
    image_path = f"../data/infographics/images/{item['image_local_name']}"  #    
    
    #    
    with Image.open(image_path) as img:
        image_width, image_height = img.size
    
    
    # OCR   
    with open(f"../data/infographics/ocr/{item['ocr_output_file']}", 'r') as ocr_file:
        ocr_data = json.load(ocr_file)
    ocr_tokens = []
    ocr_normalized_boxes = []
    
    # print(ocr_data.keys())
    if 'WORD' in ocr_data :
        for word_info in ocr_data['WORD']:
        # OCR  
            ocr_tokens.append(word_info['Text'])

        # BoundingBox     
        # 'Left', 'Top'   , 'Width', 'Height'   .
            bounding_box = word_info['Geometry']['BoundingBox']
            
            normalized_box = [
                bounding_box['Left'],
                bounding_box['Top'],
                bounding_box['Left'] + bounding_box['Width'],
                bounding_box['Top'] + bounding_box['Height']
            ]
            ocr_normalized_boxes.append(normalized_box)
    else :
        print(False)
    
    # print(ocr_tokens)
    # print(ocr_normalized_boxes)
    #  , OCR     
    #   OCR     ,      .

    data_item = {
        'question': item['question'],
        'image_id': item['image_local_name'].split('.')[0],  #  ID   
        'image_classes': None,  #    
        'extra_info': item.get('image_url'),  #    URL 
        'image_width': image_width,  #    
        'image_height': image_height,  #    
        'question_tokens': None,  # NLP    
        'question_id': item['questionId'],
        'set_name': item['data_split'],
        'image_name': item['image_local_name'],
        'image_path': f"images/{item['image_local_name']}",  #   
        'feature_path': None,  #     ,  
        # 'ocr_tokens': ocr_data.get('tokens', []),  # OCR     
        'ocr_tokens': ocr_tokens,  # OCR     
        'ocr_info': ocr_data,  #  OCR 
        'ocr_normalized_boxes': ocr_normalized_boxes,  # OCR  ,  
        'obj_normalized_boxes': None,  #   ,  
        'answers': item['answers'],
        'valid_answers': item['answers']  #  ,      
    }
    imdb_data.append(data_item)
    
    

# print(imdb_data[0]['ocr_tokens'])
# print(imdb_data[0]['question'])
# print(imdb_data[0]['image_id'])

# print(imdb_data[0]['question_id'])
# print(imdb_data[0]['image_name'])
# for item in imdb_data[0]['ocr_info']['WORD']:
#     print(item)
# NumPy  
imdb_array = np.array(imdb_data, dtype=object)
```

```python
# np.save('infographics_imdb_train.npy', imdb_array)
np.save('infographics_imdb_val.npy', imdb_array)
```
