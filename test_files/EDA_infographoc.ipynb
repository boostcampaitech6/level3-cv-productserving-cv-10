# * type distribution

```python
list1 = []
for i in data3['data']:
    for j in i['question_types']:
        if j not in list1:
            list1.append(j)
list1  # doc valid question types
```

```python
list2 = []
for i in data6['data']:
    if i['answer_type'] not in list2:
        list2.append(i['answer_type'])
list2  # info valid question types
```

```python
list3 = []  # question span
for i in data6['data']:
    if 'question span' in i['answer_type'] :
        list3.append(i)

for i in list3:
    print(i['question'], ' -> ', i['answers'], i['image_url'])
```

```python
list4 = []  # single span
for i in data6['data']:
    if 'single span' in i['answer_type'] :
        list4.append(i)

for i in list4:
    print(i['question'], ' -> ', i['answers'] , i['image_url'])
```

```python
list5 = []  # multi span
for i in data6['data']:
    if 'multi-span' in i['answer_type'] :
        list5.append(i)

for i in list5:
    print(i['question'], ' -> ', i['answers'], i['image_url'])
```

```python
list6 = []  # none extractive
for i in data6['data']:
    if 'non-extractive' in i['answer_type'] :
        list6.append(i)

for i in list6:
    print(i['question'], ' -> ', i['answers'],i['operation/reasoning'], i['image_url'])
```
