```python
import os
import json

with open("../data/qas/infographicsVQA_train_v1.0.json", "r", encoding='utf-8') as ff:
    dataset = json.load(ff)

data_concat = {}

data_concat["dataset_name"] = dataset["dataset_name"]
data_concat['dataset_version'] = dataset['dataset_version']
data_concat['dataset_split'] = dataset['dataset_split']
data = []

img_list = sorted(os.listdir('/home/lym/level3-cv-productserving-cv-10/data/images'))

for i in dataset['data']:
    if i['image_local_name'] in img_list:
        data.append(i)
data_concat['data'] = sorted(data, key = lambda x : x['image_local_name'])

with open("../data/qas/infographicsVQA_train_v1.0_custom.json", "w", encoding='utf-8') as ff:
    json.dump(data_concat, ff, indent=4)

img_list
```
